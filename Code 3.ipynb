```python
#importing libraries
import pandas as pd
import numpy as np
import datetime
```

```python
#reading dataset
data = pd.read_csv("C:\\Users\\BIPP\\Downloads\\OneDrive_1_6-20-2019\\Cleaned_Data2016-17.csv")
#filtering the required columns
df_filter = data[['Unnamed: 0', 'Commodity', 'State', 'Market Center', 'Report Date', 'Modal Prices(in Rs/quintals)']]
#finding number of rows of the dataset
len(df_filter)
```

```python
#dropping rows with null values
df_filter.dropna(inplace = True)
len(df_filter)
```

```python
#grouping the data
df = df_filter.groupby(["Commodity", "State", "Market Center"], as_index = False)
```

```python
#arranging filtered data chronolgically
df = df.apply(lambda _df: _df.sort_values('Report Date', ascending = True))
```

```python
df.head(100)
```

```python
#defining mean, standard deviation, number of data points in each group
aggregations = {'Modal Prices(in Rs/quintals)': {'mean': np.mean,  'std': np.std, 'count': 'count'}} 
```

```python
# first 5 values of the dataframe
meanstdcount.head()
```

```python
# removing multi-level index 
meanstdcount = meanstdcount.reset_index()
meanstdcount.columns = meanstdcount.columns.droplevel()
```

```python
#reviewing columns
meanstdcount.columns
```

```python
#columns are renamed
meanstdcount.columns = ['Commodity', 'State', 'Market Center', 'mean', 'std', 'count']
```

```python
meanstdcount.head()
```

```python
# merged in initial dataframe
q = df.merge(meanstdcount)
```

```python
q.head()
```

```python
#removing markets which have single data points for a given commodity
q = q[q["count"] > 1]
```

```python
#verifying that min data point are 2 in every market for every commodity
q["count"].min()
```

```python
# flag columns created to store values 
q["f1"] = np.where( ( (q["Modal Prices(in Rs/quintals)"] > (q["mean"] + q["std"])) | (q["Modal Prices(in Rs/quintals)"] < (q["mean"] - q["std"])) ) , 1, 0)
q["f2"] = np.where( ( (q["Modal Prices(in Rs/quintals)"] > (q["mean"] + (2*q["std"]))) | (q["Modal Prices(in Rs/quintals)"] < (q["mean"] - (2*q["std"])) )) , 1, 0)
q["f3"] = np.where( ( (q["Modal Prices(in Rs/quintals)"] > (q["mean"] + (3*q["std"]))) | (q["Modal Prices(in Rs/quintals)"] < (q["mean"] - (3*q["std"]))) ) , 1, 0)
q["f4"] = np.where( ( (q["Modal Prices(in Rs/quintals)"] > (q["mean"] + (4*q["std"]))) | (q["Modal Prices(in Rs/quintals)"] < (q["mean"] - (4*q["std"]))) ) , 1, 0)
q["f5"] = np.where( ( (q["Modal Prices(in Rs/quintals)"] > (q["mean"] + (5*q["std"]))) | (q["Modal Prices(in Rs/quintals)"] < (q["mean"] - (5*q["std"]))) ) , 1, 0)
q["f6"] = np.where( ( (q["Modal Prices(in Rs/quintals)"] > (q["mean"] + (6*q["std"]))) | (q["Modal Prices(in Rs/quintals)"] < (q["mean"] - (6*q["std"]))) ) , 1, 0)
```

```python
q.head()
```

```python
#the difference between prices each day is calculated to see the volatility 
#the data is not collected uniformly at equal intervals of days, therefore the difference in days is calculated 
deltadays = q["Report Date"] - q["Report Date"].shift(1)
deltaprices = q["Modal Prices(in Rs/quintals)"] - q["Modal Prices(in Rs/quintals)"].shift(1)
deltapricesbyday = deltaprices/deltadays
deltas = pd.concat([q, deltaprices, deltadays, deltapricesbyday], axis = 1)
```

```python
#moving average is calculated to see the effect of seasonality 
ma = deltapricesbyday.rolling(window = 5, center = True).mean()
```

```python
#adding columns in final data frame
final = pd.concat([deltas, ma], axis = 1)
```

```python
final.columns
```

```python
final.columns = [                  'Unnamed: 0',                    'Commodity',
                              'State',                'Market Center',
                        'Report Date', 'Modal Prices(in Rs/quintals)',
                               'mean',                          'std',
                              'count',                           'f1',
                                 'f2',                           'f3',
                                 'f4',                           'f5',
                                 'f6', 'deltaprices',
                        'deltadays',                              'deltapricebyday',
                                    'ma']
```
